## Linear Regression
Linear regression is a statistical and machine learning method used to model the relationship between one or more independent variables 
(predictors) and a continuous dependent variable (outcome). It fits the best straight line (or hyperplane in multiple dimensions) to the 
data by minimizing the residuals (the differences between actual and predicted values). Linear regression helps to understand how changes 
in predictors affect the outcome and is widely used in forecasting, trend analysis, and prediction tasks, such as predicting sales from advertising 
spend or fuel efficiency from car weight.

## Get Started 

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import sklearn.metrics as mt
import matplotlib.pyplot as plt
```

## Load The Data Set

```python
df=pd.read_csv("fifa_eda.csv")
df.head()
```

## Actual and Target Columns

```python
X=df[["Overall"]]
y=df["Potential"]
print(X)
print(y)
```

## Data Visulization

```python
plt.scatter(X, y, color='red', label='Actual')
plt.xlabel("Head Size(cm^3)")
plt.ylabel("Brain Weight(grams)")
plt.legend()
plt.show()
```

```python
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

## Train The Model

```python
model=LinearRegression()
model.fit(X,y)
```

## Slope and Intercept

```python
print("Slope: ",model.coef_)
print("Intercept: ",model.intercept_)
```

## After Data Visulization

```python
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, model.predict(X), color='red', label='Predicted')
plt.xlabel("Head Size(cm^3)")
plt.ylabel("Brain Weight(grams)")
plt.legend()
plt.show()
```

## Model Accuracy 

```python
print("Mean_Absolute_Error: ",mt.mean_absolute_error(y,model.predict(X)))
print("Mean_Square_Error: ",mt.mean_squared_error(y,model.predict(X)))
print("Root_Mean_Square_Error: ",np.sqrt(mt.mean_squared_error(y,model.predict(X))))
print("R2_Score: ",model.score(X,y))
print("R2_Score: ",mt.r2_score(y,model.predict(X)))
```
