## Polynomial Regression
Polynomial Regression is a form of linear regression in which the relationship between the independent variable x and dependent variable y is 
modeled as an nth degree polynomial. Polynomial regression fits a nonlinear relationship between the value of x and the corresponding conditional
mean of y

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
X = np.linspace(-5, 5, 50)
y = 2*X**2 + 3*X + 5 + np.random.normal(0, 5, size=X.shape)
dataset = pd.DataFrame({"x": X, "y": y})
print(dataset.head())
```

```python
plt.plot(X,y,'b.')
plt.plot(X,y,'r')
```

```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X.reshape(-1, 1))
```

```python
model = LinearRegression()
model.fit(X_poly, y)
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(X, y, 'b.', label='Data')
plt.plot(X, model.predict(poly.transform(X.reshape(-1,1))), 'r', label='Model')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Original Prediction')
plt.legend()

plt.subplot(1, 2, 2)
X_plot = np.linspace(X.min(), X.max(), 200).reshape(-1, 1)
y_plot = model.predict(poly.transform(X_plot))
plt.plot(X, y, 'b.', label='Data')
plt.plot(X_plot, y_plot, 'r', label='Polynomial Fit')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Smooth Curve')
plt.legend()
plt.tight_layout()
plt.show()
```
